## Attenuation of sound waves

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
matplotlib.rcParams.update({'font.size': 22})
```

### Useful functions

```python
def ss(x):
    return 1/(1 + x**2 )

def sqss(x):
    return np.sqrt( ss(x) )

def kk(x):
    return x*np.sqrt( (ss(x) + sqss(x) ) / 2 )

def kk_lo(x):
    return x

def kk_hi(x):
    return  np.sqrt(x / 2 )

def alpha(x):
    return x*np.sqrt( ( - ss(x) + sqss(x) ) / 2 )

def alpha_lo(x):
    return  x**2 / 2

def alpha_hi(x):
    return  kk_hi( x )

```

```python
exp_range = np.linspace(-3, 3, 200)
range= 10** ( exp_range )

exp_range_lo = np.linspace(-3, 1, 100)
range_lo= 10** ( exp_range_lo)

exp_range_hi = np.linspace(-1, 3, 100)
range_hi= 10** ( exp_range_hi)
```

```python
plt.loglog( range , ss( range ) , range , sqss( range ) )
```

```python
plt.loglog( range , kk( range ) , range , alpha( range ) )
```

```python
height=10
fig, ax = plt.subplots(figsize=(1*height, height) )
ax.loglog( range , kk( range ), 'b-', label=r'$k$, wave-vector', linewidth=3 )
ax.loglog( range_lo , kk_lo(range_lo) , 'b:', linewidth=3) #, label=r'$k$, low-$\omega$ ' )
ax.loglog( range_hi , kk_hi(range_hi) , 'k:', linewidth=3)

ax.loglog( range , alpha( range ), 'r-', label=r'$\alpha$, attenuation', linewidth=3 )
ax.loglog( range_lo , alpha_lo( range_lo ), 'r:', linewidth=3 ) #, label=r'$\alpha$, attenuation' )
#ax.loglog( range_hi , alpha_hi(range_hi) , 'r:')

ax.set_xlabel(r'$ \omega / \omega_\mathrm{c}$', fontsize='x-large')
ax.set_ylabel(r'$c \alpha / \omega_\mathrm{c}$ , $c k / \omega_\mathrm{c}$', fontsize='x-large')


#legend =  ax.legend(loc='upper left', shadow=True, fontsize='x-large')
```
