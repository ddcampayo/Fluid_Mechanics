## Attenuation of sound waves

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
```

### Useful functions

```python
def ss(x):
    return 1/(1 + x**2 )

def sqss(x):
    return np.sqrt( ss(x) )

def kk(x):
    return x*np.sqrt( (ss(x) + sqss(x) ) / 2 )

def alpha(x):
    return x*np.sqrt( ( - ss(x) + sqss(x) ) / 2 )

def alpha_lo(x):
    return  x/ 2

def alpha_hi(x):
    return  np.sqrt(1 / ( 2 * x) )

```

```python
exp_range = np.linspace(-3, 3, 200)
range= 10** ( exp_range )
```

```python
plt.loglog( range , ss( range ) , range , sqss( range ) )
```

```python
plt.loglog( range , kk( range ) , range , alpha( range ) )
```

```python
height=10
fig, ax = plt.subplots(figsize=(1*height, height) )
ax.loglog( range , kk( range ) , range , alpha( range ) , range, range, '-' )
ax.set_xlabel(r'$c \alpha / \omega_\mathrm{c}$')
```
