## Dispersion relationship for gravity-capillary waves

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
matplotlib.rcParams.update({'font.size': 22})
```

```python
def omega(k):
    return np.sqrt( k + k**3 )

def omega_lo(k):
    return np.sqrt( k  )

def omega_hi(k):
    return k**(3/2)

def cc(k):
    return np.sqrt( 1/k + k )

def cc_hi(k):
    return np.sqrt( k )

def cc_lo(k):
    return np.sqrt( 1 / k )

def cg(k):
    return ( 1 + 3*k**2 ) / ( 2 * omega(k) )

def cg_hi(k):
    return ( 3 /2 ) * cc_hi(k)

def cg_lo(k):
    return (1/2) * cc_lo(k)

```

```python
# limits set to reproduce image in wikipedia, https://en.wikipedia.org/wiki/Capillary_wave#/media/File:Dispersion_capillary.svg

lim0= np.log10( 0.01 * (2*np.pi) )
lim1= np.log10( 2 * (2*np.pi) )


kk = 10**np.linspace( lim0 , lim1 , 200)
kk_lo = kk
kk_hi = kk
#kk_lo = 10**np.linspace( lim0   , np.log10(2) , 100)
#kk_hi = 10**np.linspace( np.log10(0.5) , lim1 , 100)
```

```python
fig = plt.loglog( kk , cc(kk) , 'b-' )
```

```python
height=10
fig, ax = plt.subplots(figsize=(1*height, height) )
ax.semilogx( kk , cc( kk ), 'b-',  linewidth=3 ) # label=r'$k$, wave-vector',
ax.semilogx( kk_lo , cc_lo( kk_lo ), 'b:',  linewidth=3 )
ax.semilogx( kk_hi , cc_hi( kk_hi ), 'b:',  linewidth=3 )


ax.semilogx( kk , cg( kk ), 'r-',  linewidth=3 )
ax.semilogx( kk_lo , cg_lo( kk_lo ), 'r:',  linewidth=3 )
ax.semilogx( kk_hi , cg_hi( kk_hi ), 'r:',  linewidth=3 )

ax.set_xlabel(r'$ k / k_0 $', fontsize='x-large')
ax.set_ylabel(r'$ c \omega_0 / k_0 $', fontsize='x-large')

#t = [0.01,0.1,1,2,3,4]

#ax.set_xticks( t )
```

```python
fig.savefig('../figures/gravity_capillary_dispersion_k.eps')
```

```python
ll= 2*np.pi / kk
ll_hi= 2*np.pi / kk_hi
ll_lo= 2*np.pi / kk_lo
```

```python
height=10
fig, ax = plt.subplots(figsize=(1*height, height) )
ax.semilogx( ll , cc( kk ), 'b-',  linewidth=3 ) # label=r'$k$, wave-vector',
ax.semilogx( ll_lo , cc_lo( kk_lo ), 'b:',  linewidth=3 )
ax.semilogx( ll_hi , cc_hi( kk_hi ), 'b:',  linewidth=3 )


ax.semilogx( ll , cg( kk ), 'r-',  linewidth=3 )
ax.semilogx( ll_lo , cg_lo( kk_lo ), 'r:',  linewidth=3 )
ax.semilogx( ll_hi , cg_hi( kk_hi ), 'r:',  linewidth=3 )

ax.set_xlabel(r'$ \lambda k_0 / (2\pi) $', fontsize='x-large')
ax.set_ylabel(r'$ c \omega_0 / k_0$', fontsize='x-large')
```

```python
fig.savefig('../figures/gravity_capillary_dispersion_lambda.eps')
```
