## Dispersion relationship for gravity-capillary waves

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
matplotlib.rcParams.update({'font.size': 22})
```

```python
def omega(k):
    return np.sqrt( k + k**3 )

def omega_lo(k):
    return np.sqrt( k  )

def omega_hi(k):
    return k**(3/2)

def cc(k):
    return np.sqrt( 1/k + k )

def cc_hi(k):
    return np.sqrt( k )

def cc_lo(k):
    return np.sqrt( 1 / k )

def cg(k):
    return ( 1 + 3*k**2 ) / ( 2 * omega(k) )

def cg_hi(k):
    return ( 3 /2 ) * cc_hi(k)

def cg_lo(k):
    return (1/2) * cc_lo(k)

```

```python
kk = np.linspace( 0.1 , 5 , 200)
kk_lo = np.linspace( 0.1 , 2 , 100)
kk_hi = np.linspace( 0.8 , 5 , 100)
```

```python
fig = plt.loglog( kk , cc(kk) , 'b-' )
```

```python
height=10
fig, ax = plt.subplots(figsize=(1*height, height) )
ax.plot( kk , cc( kk ), 'b-',  linewidth=3 ) # label=r'$k$, wave-vector',
ax.plot( kk_lo , cc_lo( kk_lo ), 'b:',  linewidth=3 )
ax.plot( kk_hi , cc_hi( kk_hi ), 'b:',  linewidth=3 )


ax.plot( kk , cg( kk ), 'r-',  linewidth=3 )
ax.plot( kk_lo , cg_lo( kk_lo ), 'r:',  linewidth=3 )
ax.plot( kk_hi , cg_hi( kk_hi ), 'r:',  linewidth=3 )

ax.set_xlabel(r'$ k \ell $', fontsize='x-large')
ax.set_ylabel(r'$ c \tau / \ell $', fontsize='x-large')
```

```python
fig.savefig('../figures/gravity_capillary_dispersion_k.eps')
```

```python
ll= 2*np.pi / kk
ll_hi= 2*np.pi / kk_hi
ll_lo= 2*np.pi / kk_lo
```

```python
height=10
fig, ax = plt.subplots(figsize=(1*height, height) )
ax.plot( ll , cc( kk ), 'b-',  linewidth=3 ) # label=r'$k$, wave-vector',
ax.plot( ll_lo , cc_lo( kk_lo ), 'b:',  linewidth=3 )
ax.plot( ll_hi , cc_hi( kk_hi ), 'b:',  linewidth=3 )


ax.plot( ll , cg( kk ), 'r-',  linewidth=3 )
ax.plot( ll_lo , cg_lo( kk_lo ), 'r:',  linewidth=3 )
ax.plot( ll_hi , cg_hi( kk_hi ), 'r:',  linewidth=3 )

ax.set_xlabel(r'$ \lambda / \ell $', fontsize='x-large')
ax.set_ylabel(r'$ c \tau / \ell $', fontsize='x-large')
```

```python
fig.savefig('../figures/gravity_capillary_dispersion_lambda.eps')
```
