## Taylor-Green vortices

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
matplotlib.rcParams.update({'font.size': 22})
```

```python
def cc(x ):
    return np.cos( 2* np.pi * x )

def ss(x ):
    return np.sin( 2* np.pi * x )

def press(x , y , t ):
    cx = cc(2*x)
    cy = cc(2*y)
    amp = -np.exp(-4*t) / 4
    return amp*(cx + cy)

def u (x,y,t):
    cx = cc(x); sx=ss(x)
    cy = cc(y); sy=ss(y)
    amp = np.exp(-2*t)
    return amp*(cx * sy) , -amp*(sx * cy)

```

```python
deltax=0.01
lim = 0.5
nn = int(2*lim/deltax)
x = np.linspace( -lim , lim , nn )
y = np.linspace( -lim , lim , nn )
X, Y = np.meshgrid(x, y)
P = press ( X  ,  Y , 0 )

#coarser for the velocity field
deltax2=0.05

nn2 = int(2*lim/deltax2)
x2 = np.linspace( -lim , lim , nn2 )
y2 = np.linspace( -lim , lim , nn2 )
X2, Y2 = np.meshgrid(x2, y2)


UX , UY = u ( X2  ,  Y2 , 0 )
```

```python
height=10
fig, ax = plt.subplots(figsize=(1*height, height) )
CS2 = ax.contour(X, Y, P )#, levels= np.arange(-2, 2, 0.1) ) 
#ax.clabel(CS2, inline=1, fontsize=20)
ax.set_title('pressure')
#ax2.set_ylim(0 ,lim)
#ax.set_aspect( 1 )
```

```python
height=10
fig, ax = plt.subplots(figsize=(1*height, height) )
CS2 = ax.contourf(X, Y, P )
qu = ax.quiver(X2, Y2, UX, UY )
ax.set_title('pressure, velocity')
ax.set_xlabel(r'$x/L$', fontsize='x-large')
ax.set_ylabel(r'$y/L$', fontsize='x-large')
```
