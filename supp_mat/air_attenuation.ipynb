```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
data= scipy.loadtxt( 'air_att_coeff.dat' )
```

```python
f=data[:,0]
```

```python
al=data[:,1]
```

```python
slope, incercept, rvalue, pvalue, stderr = st.linregress( f,al)
```

```python
plt.plot(f,al,'o', f, incercept + slope * f )
```

```python
slope, incercept, rvalue, pvalue, stderr = st.linregress(f**2 , al)
```

```python
plt.plot(f,al,'o', f, incercept + slope * f**2 )
```

```python
plt.loglog(f,al,'o' , f, incercept + slope * f**2 )
```

```python
plt.loglog(  f, incercept + slope * f**2 )
```
