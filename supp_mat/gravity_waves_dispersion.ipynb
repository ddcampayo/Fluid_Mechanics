## Dispersion relation for gravity waves

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
matplotlib.rcParams.update({'font.size': 22})
```

### Useful functions

```python
def cc(k):
    return np.sqrt(np.tanh(k) / k)

def cc_hi(k):
    return 1/np.sqrt(k )

def cc_lo(k):
    return 1+0*k # trick!

def cg(k):
    return cc(k)/2 * (  1+ k * ( 1- np.tanh(k)**2 ) / np.tanh(k)  )

def cg_hi(k):
    return cc_hi(k)/2

def cg_lo(k):
    return cc_lo(k)

```

```python
exp_range = np.linspace(-2, 2, 200)
range= 10** ( exp_range )

exp_range_lo = np.linspace(-2, 0.5, 100)
range_lo= 10** ( exp_range_lo)

exp_range_hi = np.linspace(-0.5, 2, 100)
range_hi= 10** ( exp_range_hi)
```

```python
plt.semilogx( range_hi , cg_hi( range_hi ) ) #, range , cg( range ) )
```

```python
height=10
fig, ax = plt.subplots(figsize=(1*height, height) )
ax.semilogx( range , cc( range ), 'b-', linewidth=3 )
ax.semilogx( range_lo , cc_lo(range_lo) , 'b:', linewidth=3) #, label=r'$k$, low-$\omega$ ' )
ax.semilogx( range_hi , cc_hi(range_hi) , 'b:', linewidth=3)

ax.semilogx( range , cg( range ), 'r-',  linewidth=3 )
#ax.semilogx( range_lo , cg_lo( range_lo ), 'r:', linewidth=3 ) #, label=r'$\alpha$, attenuation' )
ax.semilogx( range_hi , cg_hi( range_hi ), 'r:', linewidth=3 )

ax.set_xlabel(r'$ k h $', fontsize='x-large')
ax.set_ylabel(r'$c / \sqrt{g h}  $', fontsize='x-large')


#legend =  ax.legend(loc='upper left', shadow=True, fontsize='x-large')
```

```python
height=10
fig, ax = plt.subplots(figsize=(1.5*height, height) )
ax.semilogx( range / (2*np.pi) , cc( range ), 'b-', linewidth=3 )
ax.semilogx( range_lo/ (2*np.pi) , cc_lo(range_lo) , 'b:', linewidth=3) #, label=r'$k$, low-$\omega$ ' )
ax.semilogx( range_hi/ (2*np.pi) , cc_hi(range_hi) , 'b:', linewidth=3)

ax.semilogx( range/ (2*np.pi) , cg( range ), 'r-',  linewidth=3 )
#ax.semilogx( range_lo , cg_lo( range_lo ), 'r:', linewidth=3 ) #, label=r'$\alpha$, attenuation' )
ax.semilogx( range_hi/ (2*np.pi) , cg_hi( range_hi ), 'r:', linewidth=3 )

ax.set_xlabel(r'$  h / \lambda $', fontsize='x-large')
ax.set_ylabel(r'$c / \sqrt{g h}  $', fontsize='x-large')
```
