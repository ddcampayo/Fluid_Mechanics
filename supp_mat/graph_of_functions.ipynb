## Simple function drawing

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
matplotlib.rcParams.update({'font.size': 22})
```

### Useful functions

```python
def ecos(x):
    return np.exp(-x)*np.cos(x)

def qq2(q):
    return 1/(q*(1+q))

def qq2_lo(q):
    return 1/q

def qq2_hi(q):
    return 1/q**2

```

```python
range = np.linspace(0, 10, 200)

exp_range = np.linspace(-3, 3, 200)
erange= 10** ( exp_range )

exp_range_lo = np.linspace(-3, 1, 100)
erange_lo= 10** ( exp_range_lo)

exp_range_hi = np.linspace(-1, 3, 100)
erange_hi= 10** ( exp_range_hi)
```

```python
height=10
fig, ax = plt.subplots(figsize=(1.5*height, height) )
ax.plot( range , ecos( range ), 'b-', linewidth=3 )

#ax.loglog( range_lo , kk_lo(range_lo) , 'b:', linewidth=3) #, label=r'$k$, low-$\omega$ ' )

ax.set_xlabel(r'$ x $', fontsize='x-large')
ax.set_ylabel(r'$ e^{-x} \cos(x) $', fontsize='x-large')


#legend =  ax.legend(loc='upper left', shadow=True, fontsize='x-large')
```

```python
fig.savefig('../figures/expcos.eps')
```

```python
height=10
fig, ax = plt.subplots(figsize=(1.5*height, height) )
ax.loglog( erange , qq2(erange) , 'b', linewidth=3) #, label=r'$k$, low-$\omega$ ' )
ax.loglog( erange_lo , qq2_lo(erange_lo) , 'b:', linewidth=3)
ax.loglog( erange_hi , qq2_hi(erange_hi) , 'b:', linewidth=3)


#ax.loglog( erange , qq2(erange) , 'b:', linewidth=3) #, label=r'$k$, low-$\omega$ ' )

ax.set_xlabel(r'$ q \xi $', fontsize='x-large')
ax.set_ylabel(r'$ 1 \over (q\xi) ^2  + q \xi $', fontsize='x-large')


#legend =  ax.legend(loc='upper left', shadow=True, fontsize='x-large')
```
